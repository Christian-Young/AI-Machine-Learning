Problem 3

What does an average MNIST digit look like? For each i=0,1,...,9, compute the average of digit i and display it. More precisely, you have add all the images of the digit i together and divide it by the number of times the digit i occurs in the data set.

Use tf.keras.datasets to load the MNIST digits dataset.

#The intention of Problem 3 is to display the average MNIST digit based on the keras datasets. The following notebook is used as a reference. [MNIST dataset notebook](https://colab.research.google.com/drive/1HDZB0sEjhd0sdTFNCmJXvB8hYnE9KBM7)

```
%tensorflow_version 2.x
import tensorflow as tf
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
(train_images, train_labels), (test_images, test_labels) = tf.keras.datasets.mnist.load_data()
```

```
# Initiate 10 x 28 x 28 zeroed arrays
train = np.zeros([10, 28, 28])

# Sum up images and labels to obtain an average
for i in range(len(train_images)):
  train[train_labels[i]] = train[train_labels[i]] + train_images[i]
```

```
# Display each digit based on the averages
plt.figure(figsize = (12, 12))
for i in range(10):
  ax = plt.subplot(4, 4, i + 1)
  plt.imshow(train[i])
  plt.title(i)
  plt.axis('off')
```
