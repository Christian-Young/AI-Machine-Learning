#The intention of Problem 1 is to Find three simple bivariate functions f1, f2, and f3 and three points p1, p2, p3 such that p1 is a minimum of f1 p2 is a maximum of f2 p3 is neither a minimum nor maximum of f3, but the gradient of f3 at p3 is the zero vector. The following notebooks provided by Dr. Wocjan were referenced: [3-Dimensional Plotting](https://colab.research.google.com/drive/13x9hKVoc-A295zxRSSDOELVqK8jCiQsO), [Density and Contour Plots](https://colab.research.google.com/drive/1dASWY0YX9pPuwjk95RaBGiArycd8KG22)

###Bivariate function 2 will be defined as:
\begin{equation*}
f(a, b) = \sqrt a * \sqrt b
\end{equation*}

The following is a contour plot of the function:

```
import numpy as np
%matplotlib inline
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt

# Simple function defined as sqrt(a) * sqrt(b). The local min will be (0, 0)
def biv1(a, b):
  return np.sqrt(a) * np.sqrt(b)

xline = np.linspace(0, 100, 100)
yline = np.linspace(0, 100, 100)
zline = biv1(xline, yline)
x, y = np.meshgrid(xline, yline)
z = biv1(x, y)
plt.style.use('dark_background')
plt.contour(x, y, z, 30, cmap = 'Blues')
plt.scatter(0, 0)
plt.show()
```

###Density plot:

```
plt.contourf(x, y, z, 40, cmap = 'Blues')
plt.colorbar()
plt.scatter(0, 0)
plt.show()
```

### 3-Dimensional plot:

```
fig = plt.figure()
ax = plt.axes(projection = '3d')
ax.contour3D(x, y, z, 50)
ax.scatter(0, 0, 0, c = 'Blue')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.view_init(50, 90)
```

###Bivariate function 2 will be defined as:
\begin{equation*}
f(a, b) = -\sqrt a - b
\end{equation*}

The following is a contour plot of the function:

```
# Function defined as: -sqrt(a) - b. The local max will also be (0, 0).
def biv2(a, b):
  return -np.sqrt(a) - b

xline = np.linspace(0, 10, 100)
yline = np.linspace(0, 10, 100)
zline = biv2(xline, yline)
x, y = np.meshgrid(xline, yline)
z = biv2(x, y)
plt.style.use('dark_background')
plt.contour(x, y, z, 30, colors = 'blue')
plt.scatter(0, 0)
plt.show()
```

###Density plot:

```
plt.contourf(x, y, z, 40, cmap = 'Blues')
plt.colorbar()
plt.scatter(0, 0)
plt.show()
```

###3-Dimensional plot:

```
fig = plt.figure()
ax = plt.axes(projection = '3d')
ax.plot_surface(x, y, z)
ax.scatter(0, 0, 0, c = 'Blue')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.view_init(40, 20)
```

###Bivariate function 3 will be defined as:
\begin{equation*}
f(a, b) = cos(a) / b
\end{equation*}

The following is a contour plot of the function:

```
# Defined as: cos(a)/b. (0, 0) is neither a local max or min for the function.
def biv3(a, b):
  return np.cos(a)/b

xline = np.linspace(-10, 10, 100)
yline = np.linspace(-10, 10, 100)
zline = biv3(xline, yline)
x, y = np.meshgrid(xline, yline)
z = biv3(x, y)
plt.style.use('dark_background')
plt.contour(x, y, z, 100, colors = 'blue')
plt.scatter(0,0)
plt.show()
```

###Density plot:

```
plt.contourf(x, y, z, 100, cmap = 'Blues')
plt.colorbar()
plt.scatter(0, 0)
plt.show()
```

###3-Dimensional plot:

```
fig = plt.figure()
ax = plt.axes(projection = '3d')
ax.scatter(0, 0, 0, c = 'Blue', marker = 'o')
ax.plot_surface(x, y, z)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.view_init(30, 60)
```
